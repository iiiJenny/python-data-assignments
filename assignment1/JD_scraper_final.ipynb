```python
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException


#抓取每个unit的数据

def scrap_each_phone_info():
    all_phones = browser.find_elements_by_css_selector("li.gl-item")
    phone_info = []
    failed = []
    for phone in all_phones:
        info = {} 
        try:
            info['name'] = phone.find_element_by_class_name("p-name-type-2").find_element_by_tag_name("a").find_element_by_tag_name("em").text
            info['price'] = phone.find_element_by_class_name("p-price").find_element_by_tag_name("i").text
            info['comment'] = phone.find_element_by_class_name("p-commit").find_elements_by_tag_name("a")[-1].text
            info['shop'] = phone.find_element_by_class_name("p-shop").find_element_by_tag_name("a").text
            other_info = phone.find_element_by_class_name("p-icons").find_elements_by_tag_name("i")
            coupon = []
            for i in other_info:
                coupon.append(i.text)
            info['coupon'] = coupon
            phone_info.append(info)
            
        except NoSuchElementException as exception:
                print("Element not found")
                failed.append(info['name'])
                
    return phone_info
```

```python
url = "https://www.jd.com/index.html"
browser = webdriver.Chrome()
# browser.wait = WebDriverWait(browser, 5)
browser.get(url)
keywords = browser.find_element_by_id("key").send_keys("手机\n")
```

```python
data = []
for i in range(0,10):
    time.sleep(5) 
    element = browser.find_element_by_class_name("pn-next")
    browser.execute_script("return arguments[0].scrollIntoView();", element)
    data.extend(scrap_each_phone_info())
    element = browser.find_element_by_class_name("pn-next")
    element.click()
```
