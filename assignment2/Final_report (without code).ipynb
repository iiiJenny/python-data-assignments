# Peer into Job Market for Media Students

### 1. Background
With the transformation of economic structure, the current contradiction between the polarization of the occupational structure in China is prominent. The "China Employment Market Report" released by the China Employment Research Institute of Renmin University of China shows that talents in emerging industries are in short supply, and traditional industries are more competitive. For example, in the new economic industries such as the Internet, e-commerce, securities and transportation, the talents are in short supply under the dual promotion of the market and policies.

Therefore, our group decide to explore the current employment market, and analyze the employment situation for media graduates.

#### (2) On the perspective of job seekers
After collecting the data from the report released by "看准网", we can see that the IT industry here, with the proportion of 43% ranks first in the list.

In the job market, companies and job seekers all play a very important role, so, in this part we scrap the relevant data from the perspective of job seekers. From the perspective of the job seeker, the industries (top10) that graduates are most concerned about are the field of Internet, finance, and real estate industries. Among them, the Internet accounted for the largest proportion (43.3%) and was much higher than other industries, that is, the Internet industry received the highest attention.

In addition to the industry, the choice of city is also important for most job seekers.

According to the report released by the pay network, the top 10 cities for graduates to choose to work are: Guangzhou, Chengdu, Beijing, Chongqing, Shenzhen, Zhengzhou, Shanghai, Changsha, Hangzhou, Guiyang.
The graphs here, we can see that most of the cities that they choose, like Guangzhou, Chengdu, Shanghai, are the first-tier cities. So it seems that graduates are more eager to work in first-tier cities.

In order to explore the reasons for the popularity of the Internet industry, our team has scraped comments on the Internet industry and made word clouds. It can be found that the words “development/success/success/high salary/foreground” are frequently appeared in the word cloud, which explains the reason why so many job seekers are concerned about IT industry.

In addition to the Internet, our group feels that it is necessary to explore the reason why the field of media is not so popular in all industries? For example, "resignation / bad / cervical spondylosis / coronary heart disease / stomach disease / lumbar vertebrae", etc., it can be seen that the work in the media industry is not easy.

So in this part, we mentioned about the context of fresh graduates’ employment. It can be seen that entering into the IT industry becomes the mainstream of the fresh graduates’ employment.
But for those people who strongly tend to work in the media industry like us, it still deserve to learn about the current situation in the Media industry.


### 2. Companies for media students

#### (1) Internet companies & media jobs from internet company
Based on our understanding of the background of employment market, we hope to further explore at the company level. First of all, most job seekers want to work in a "good" company. Generally speaking, a good company represent a large market share, a good salary, a good learning platform or a potential promotion space.

Take “看准网” (www.kanzhun.com) as an example. According to the hot search rankings of the company in Kanzhun website, we have scraped the top 90 companies that are most popular for job seekers, we show the top 20 here. Most companies in top10 are familiar: BAT, Huawei, Jindong, Foxconn, etc., belong to the Internet or e-commerce field.


```python
#原始数据
df = pd.read_csv("clean_kanzhun_hot_corp_info0.csv")
del df['Unnamed: 0']
len(df) #90
df.head(10) #显示部分数据
```

Looking at all the hot companies (total: 90), we divide them according to their fields, and we can find that these popular companies can be mainly divided into eight fields: Internet, banking, communications, catering, e-commerce, etc., among which Internet companies still account for the highest (20%), followed by the bank (6.92%). This pie chart illustrates the large demand for new economic industries (such as the Internet, communications, e-commerce, etc.).

```python
# 拆分type信息，并放在同一列上
types = df['type'].str.split('/', expand=True).stack().reset_index(level=1, drop=True)#.rename('type')
types.index = range(1,len(types) + 1) #重新index
len(types) #130

types0 = types[:]
a = ['计算机软件 ','IT服务','互联网 ','网络设备 ','门户网站']
b = 0
for i in range(1,131):
    if types0[i] in a:
        types0[i] = '互联网公司'
        b = b + 1
b # 30
types0.value_counts()[:10]#.sum()
```

Secondly, we conducted descriptive statistics on 90 popular companies in terms of salary. The results show that the average monthly salary of 90 hot companies is 6315RMB, the highest is 14520RMB (Microsoft China). In contrast, the top monthly salary of popular companies top20 is 10045RMB, which is higher than the average salary of all popular companies by nearly 4000RMB. It can be seen that the salary level is also an important criterion for measuring hot companies.


```python
len(df['salary']) #90
df['new_salary'] = df['salary'].str.lstrip("平均工资  ￥")

for i in range(0,90):
    df['new_salary'][i] = int(df['new_salary'][i])
df2 = df.sort_values(by='new_salary', ascending=False)[:20]
df2[:10]
```
